```python
vc = cv2.VideoCapture(r'D:\Jupyter\OpenCV\.mp4')
```

```python
if vc.isOpened():
    open, frame = vc.read()
else:
    open = False
```

```python
while open:
    ret, frame = vc.read()
    if frame is None:
        break
    if ret == True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('reslut', gray)
        if cv2.waitKey(25) & 0xFF == 27:
            break
vc.release()
cv2.destroyAllWindows()
```
