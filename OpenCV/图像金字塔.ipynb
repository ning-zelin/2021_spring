* ## API

* cv.pyrUp(img)
* cv.pyrDown(img)

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread( r'D:\DeepLearning\Data\antbees(2)\train\ants\0013035.jpg')
```

```python
up_img = cv.pyrUp(img)
print(up_img.shape)
plt.imshow(up_img[:,:,::-1])
```

```python
up_img = cv.pyrDown(img)
print(up_img.shape)
plt.imshow(up_img[:,:,::-1])
```
