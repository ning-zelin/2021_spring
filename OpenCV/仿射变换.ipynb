## 仿射变换

* 通过对图像的点作映射, 线平行, 比例关系不会变化

>指定矩阵, 2*3 线性变换 + 平移


> M = cv.getAffineTransform(pts1, pts2)
> cv.warpAffine(img, M, (cols, rows))

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread( r'D:\DeepLearning\Data\antbees(2)\train\ants\0013035.jpg')
```

```python
cv.imshow('img',img)
cv.waitKey(0)
```

```python
rows, cols = img.shape[:2]
print(rows, cols)
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[100,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1, pts2)
```

```python
M
```

```python
dst = cv.warpAffine(img, M, (cols, rows))
```

```python
def im_show(img):
    cv.imshow('img', img)
    cv.waitKey(0)
```

```python
im_show(dst)
```

```python
pts1 = np.float32([[50,50],[100,50],[200, 100]])
pts2 = np.float32([[100,30],[21,40],[100,200]])
M = cv.getAffineTransform(pts1, pts2)
img_af = cv.warpAffine(img, M, (cols, rows))
```

```python
im_show(img_af)
```
