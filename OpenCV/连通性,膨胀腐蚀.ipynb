## 连通性
* 连通  : 位置相邻, 灰度值相近
* 四连通: 位置相邻, 灰度值相近, 在四邻域里
* 四连通: 位置相邻, 灰度值相近, 在爸邻域里
* M 连通: 位置相邻, 灰度值相近, 在四邻域/D邻域里, 连个四邻域的交集为空
