```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread( r'D:\DeepLearning\Data\antbees(2)\train\ants\0013035.jpg')
rows, cols = img.shape[:2]
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[56,75],[468,52],[48,387],[339,390]])
T = cv.getPerspectiveTransform(pts1, pts2)
print(T)
```

```python
img_Pers = cv.warpPerspective(img, T, (rows, cols))
cv.imshow('m', img_Pers)
cv.waitKey(0)
```
